# 平面划分

```python
#产生新画布
fig=plt.figure()
#获取当前画布
figa=plt.gca()

#产生100个点
N=100
xn=np.random.rand(N,2)
x=np.linspace(0,1)

#随机产生一条线
a=np.random.rand()
b=np.random.rand()
f=lambda x:a*x+b
plt.plot(x,f(x),'r')

#线性分割前面产生的点
yn=np.zeros([N,1])

for i in range(N):
    if(f(xn[i,0]))>=xn[i,1]:
        #point is below line
        yn[i]=1
        plt.plot(xn[i,0],xn[i,1],'bo',markersize=12)
    else:
        #point is above line
        yn[i]=-1
        plt.plot(xn[i,0],xn[i,1],'go',markersize=12)

```

# 感知器简单实现

```python
def perceptron(xn,yn,MaxIter=1000,a=0.1,w=np.zeros(3)):
    N=xn.shape[0]
    #生成超平面
    f=lambda x:np.sign(w[0]*1+w[1]*x[0]+w[2]*x[1])
    for _ in range(MaxIter):
        i=np.random.randint(N)
        if(yn[i]!=f(xn[i,:])):
            w[0]=w[0]+yn[i]*1*a
            w[1]=w[1]+yn[i]*xn[i,0]*a
            w[2]=w[2]+yn[i]*xn[i,1]*a
    return w
   
```

# 实际应用

```python
w=perceptron(xn,yn)
    
#利用权重w计算y=a*x+b中的a和b
bnew=-w[0]/w[2]
anew=-w[1]/w[2]
y=lambda x:anew*x+bnew

#分割颜色
sep_color=yn/2.0
plt.figure()
figa=plt.gca()

plt.scatter(xn[:,0],xn[:,1],c=sep_color.flatten(),s=50)
plt.plot(x,y(x),'b--',label='感知器分类结果')
plt.plot(x,f(x),'r',label='原始分类曲线')
plt.legend()
plt.title('原始曲线与感知器近似结果比较')

    
  
```
